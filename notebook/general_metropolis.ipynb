```python
import learn_metropolis as lm
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_simulations = 2000
z, N = 14, 20
sd = 1
theta = 0.01
hist = np.zeros(num_simulations)
hist[0] = theta

likelihood = lm.gen_likelihood_function(z, N)
proposed = lm.gen_proposed_function(sd)
```

```python
for idx in range(1, num_simulations):
    current_param = hist[idx-1]
    hist[idx] = lm.metropolis_stepping(current_param, likelihood, proposed)
```

```python
plt.plot(hist, range(num_simulations))
```
